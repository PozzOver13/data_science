## Reshaping
stack = This “rotates” or pivots from the columns in the data to the rows (long or stacked format)

unstack = This pivots from the rows into the columns

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                    index=pd.Index(['Ohio', 'Colorado'], name='state'),
                    columns=pd.Index(['one', 'two', 'three'],
                    name='number'))
data
```

```python
results = data.stack()
results
```

```python
results.unstack()
```

```python
results.unstack(0)
```
